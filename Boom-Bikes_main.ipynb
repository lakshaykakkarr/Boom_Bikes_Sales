# BOOM BIKES Linear Regression Assignment

#### Problem Statement: 

A US bike-sharing provider BoomBikes has recently suffered considerable dips in their revenues due to the ongoing Corona pandemic. They have contracted a consulting company to understand the factors on which the demand for these shared bikes depends. Specifically, they want to understand the factors affecting the demand for these shared bikes in the American market.

The company wants to know:

- Which variables are significant in predicting the demand for shared bikes.

- Model the demand for shared bikes with the available independent variables.

```python
# dropping unneccessary columns
# 'instant' unique column, meaning each value is different for each row, not important for the model
# 'dteday' date column, information already present in 'year' and 'month' columns thus, not important
# 'casual' and 'registered' together form the 'count' column, which is our target variable
# 'temp' temperature column, another column with 'atemp' adjusted temperature is present, both are highly correlated

bike.drop(labels = ['instant', 'dteday', 'casual', 'registered', 'temp'], axis=1, inplace = True)
bike.head()
```

```python
# renaming columns just to get a clean feel of the data
# 'yr' to 'year'
# 'mnth' to 'month'
# 'hum' to 'humidity'
# 'cnt' to 'count'
# 'atemp' to 'real_temp'

bike.rename(columns = {'yr': 'year', 'mnth': 'month', 'hum': 'humidity', 'cnt': 'count', 'atemp': 'real_temp'}, inplace = True)
bike.head()
```

```python
# redefining categorical columns according to data dictionary to get appropriate dummy variables
# 'season' [1: spring, 2: summer, 3: fall, 4: winter]
# 'year' [0: 2018, 1: 2019]
# 'month' [1: january, ... , 12: december]
# 'weathersit' [1: clear, 2: mist_cloudy, 3: light_snow]

bike['season'] = bike['season'].map({1: 'spring', 2: 'summer', 3: 'fall', 4: 'winter'})
bike['month'] = bike['month'].map({1: 'january', 2: 'february', 3: 'march', 4: 'april', 5:'may', 6: 'june', 7: 'july', 8: 'august', 9: 'september', 10: 'october', 11: 'november', 12: 'december'})
bike['weekday'] = bike['weekday'].map({0: 'sunday', 1: 'monday', 2: 'tuesday', 3:  'wednesday', 4: 'thursday', 5: 'friday', 6: 'saturday'})
bike['weathersit'] = bike['weathersit'].map({1: 'clear', 2: 'mist_cloudy', 3: 'light_snow',  4: 'heavy_rain'})
bike.head()

# bike['year'] = bike['year'].map({0: '2018', 1:'2019'})
# bike['holiday'] = bike['holiday'].map({0 :'no', 1: 'yes'})
# bike['workingday'] = bike['workingday'].map({0: 'no', 1: 'yes'})
```

## Step 3: Data Preparation

#### Creating Dummy Variables

```python
# creating (k-1) dummy variables for the categorical columns using get_dummies()
# checking the obtained dummy vars

d1 = pd.get_dummies(bike['season'], drop_first = True)
d2 = pd.get_dummies(bike['month'], drop_first = True)
d3 = pd.get_dummies(bike['weekday'], drop_first = True)
d4 = pd.get_dummies(bike['weathersit'], drop_first = True)
```

```python
# checking the obtained dummy vars
```

```python
d2.head() # months
```

```python
d3.head() # weekdays
```

```python
d4.head() # weathersits
```

```python
# concatinating all the dummy variable columns to get original dataframe
# removing the original categorical columns

bike = pd.concat([bike, d1, d2, d3, d4], axis = 1)
remcol = ['season', 'month', 'weekday', 'weathersit']
bike.drop(bike[remcol], axis = 1, inplace=True)
bike.head()
```

```python
# checking the dataset before developing model

bike.info()
```

## Step 4: Splitting the Data into Training and Testing Sets


```python
# splitting the dataset into train and test using sklearn

from sklearn.model_selection import train_test_split
bike_train, bike_test = train_test_split(bike, train_size = 0.7, test_size = 0.3, random_state = 100)
# random_state = 100, so as to get same test and train data split
```

#### Rescaling the Features 

Two ways:
1. Min-Max scaling 
2. Standardisation (mean-0, sigma-1) 

We will use MinMax scaling

```python
# scaling continuous variables using MinMaxScaler()
# only continuous columns except the dummy variable columns and binary columns are to be scaled

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# 'real_temp', 'humidity' and 'windspeed' scaling using fit_transform()

concol = ['real_temp', 'humidity', 'windspeed']
bike_train[concol] = scaler.fit_transform(bike_train[concol])
bike_train.head()
```

```python
# checking the scaled columns
# values should be between 0-1, max = 1

bike_train.describe()
```

#### Dividing into X and Y sets for the model building

```python
# creating target(y) and  predictors(X) datasets

y_train = bike_train.pop('count')
X_train = bike_train
# print(y_train)
# print(X_train)
```

## Step 5: Building a linear model


#### RFE
Recursive Feature Elimination

```python
# feature selection using RFE technique (Recursive Feature Elimination)
# top 15 variables to be selected

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train, y_train)
rfe =  RFE(lm, 15)
rfe = rfe.fit(X_train, y_train)
```

```python
# checking the selected features and their rankings

list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
# keeping only the RFE selected columns and removing others from the predictor dataset
```

```python
rfecol = X_train.columns[rfe.support_]
rfecol
```

```python
X_train.columns[~rfe.support_]
```

#### Building model using statsmodel, for the detailed statistics

```python
X_train_rfe = X_train[rfecol]
X_train_rfe.head()
```

#### Adding all the variables to the model


```python
import statsmodels.api as sm

X_train_lm = sm.add_constant(X_train_rfe)  # adding constant variable
lr = sm.OLS(y_train, X_train_lm).fit()     # running the model
print(lr.summary())                        # checking summary of the model
```

#### Checking VIF

Variance Inflation Factor or VIF, gives a basic quantitative idea about how much the feature variables are correlated with each other. It is an extremely important parameter to test the linear model. The formula for calculating `VIF` is:

### $ VIF_i = \frac{1}{1 - {R_i}^2} $

```python
# checking VIF for all variables in the model to study the multicollinearity in the model
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

##### Insights:
- variables with high p-values (>= 0.05) and high VIFs (>5) to be removed

#### Dropping the variable and updating the model


```python
# building model with all available variables and dropping one-by-one 
```

```python
# lr_1 without 'saturday'

X = X.drop('saturday', 1)

X_train_lm = sm.add_constant(X)
lr_1 = sm.OLS(y_train, X_train_lm).fit()

print(lr_1.summary())
print('\n'*2)
print('VIF Check')

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
print(vif)
```

## Step 7: Making Predictions Using the Final Model


#### Applying the scaling on the test sets

```python
# scaling the test data using transform()

concol = ['real_temp', 'humidity','windspeed']
bike_test[concol] = scaler.transform(bike_test[concol])
bike_test.head()
```

```python
bike_test.describe()
```

#### Dividing into X_test and y_test

```python
# creating target and predictor sets for test data

y_test = bike_test.pop('count')
X_test = bike_test
```

```python
X_test_lm.head()
```

```python
# making predictions using the lr_5 model

y_pred_lm = lr_5.predict(X_test_lm)
```

## Step 8: Model Evaluation

plotting the graph for actual versus predicted values

```python
# model evaluation
# plotting y_test and y_pred to understand the spread

fig = plt.figure()
plt.scatter(y_test, y_pred_lm)
plt.title('y_test vs y_pred', fontsize = 20)
plt.xlabel('y_test', fontsize = 18)          # X-label
plt.ylabel('y_pred', fontsize = 16)          # Y-label
plt.show()
```

#### Checking the R-squared on the test set

```python
# checking the R-squared on the test set
r_squared = r2_score(y_test, y_pred_lm)
r_squared
```
